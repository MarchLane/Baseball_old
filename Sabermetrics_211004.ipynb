### OPS

#### 출루율 (안타+볼넷+몸에 맞은 공) / (타수+볼넷+몸에 맞은 공+희생플라이)
(HIT+BB+HP)/(AB+BB+HP+SF)

```python
bat_19
```

```python
OBP_list = []
for i in range(0, len(bat_19)) :
    up = bat_19.iloc[i, 4] + bat_19.iloc[i, 12] + bat_19.iloc[i, 14]
    down = bat_19.iloc[i, 1] + bat_19.iloc[i, 12] + bat_19.iloc[i, 14] + bat_19.iloc[i, 11]
    OBP = up/down
    OBP_list.append(OBP)
```

```python
bat_19['OBP'] = OBP_list
bat_19
```

#### 장타율 (단타 수 + 2루타 수*2 + 3루타 수*3 + 홈런 수*4)/타수

```python
SLG_list = []
for j in range(0, len(bat_19)) :
    hit1 = bat_19.iloc[j, 4] - bat_19.iloc[j, 5] - bat_19.iloc[j, 6] - bat_19.iloc[j, 7]
    hit2 = bat_19.iloc[j, 5] * 2
    hit3 = bat_19.iloc[j, 6] * 3
    hit4 = bat_19.iloc[j, 7] * 4
    
    up = hit1 + hit2 + hit3 + hit4
    down = bat_19.iloc[j, 1]
    
    SLG = up/down
    SLG_list.append(SLG)
```

```python
bat_19['SLG'] = SLG_list
bat_19
```

#### OPS = 출루율 + 장타율

```python
bat_19['OPS'] = [k + l for k, l in zip(bat_19['OBP'], bat_19['SLG'])]
bat_19
```

### SECA (Secondary Average)
SECA = (BB + (TB - H) + (SB -CS)) / AB

```python
SecA는 루타에서 단타를 빼 추가 진루를 얼마나 얻어냈는지 보는 스탯입니다. 
타율은 장타가 구분되지 않고 볼넷이 포함되지 않기 때문에 타율로  평가할 수 없는 부분인 장타력과 
선구안을 보는 것입니다. 분자에 볼넷이 들어가는데 분모에는 볼넷이 포함되지 않는 문제가 있습니다.
단타가 빠지기 때문에 SecA 만으로는 종합적인 타격능력을 평가할 수는 없고 타율의 보조적인 역할을 합니다.
리그평균 SecA는 0.200 ~ 0.250 정도입니다.
```

```python
bat_19
```

```python
B_list = [bat_19.iloc[n, 5] * 2 + bat_19.iloc[n, 6] * 3 + bat_19.iloc[n, 7] * 4 
          for n in range(0, len(bat_19))]
SECA_list = [(bat_19.iloc[o, 12] + B_list[o] + bat_19.iloc[o, 8] - bat_19.iloc[o, 9]) / 
            bat_19.iloc[o, 1] for o in range(0, len(bat_19))]
bat_19['SECA'] = SECA_list
bat_19
```

### TA
TA shows a batter’s offensive contribution. The higher the number, the more contributions and the fewer negative contributions.

```python
TA = (Total Bases + Hit by Pitch + Walks + Stolen Base) ÷ 
(At Bats – Hits + Caught Stealing + Ground Into Double Play)
```

```python
bat_19
```

```python
TB_list = [bat_19.iloc[p, 4] - (bat_19.iloc[p, 5] + bat_19.iloc[p, 6] + bat_19.iloc[p, 7])
           + bat_19.iloc[p, 5] * 2 + bat_19.iloc[p, 6] * 3 + bat_19.iloc[p, 7] * 4
           for p in range(0, len(bat_19))]
TA_list = [(TB_list[q] + bat_19.iloc[q, 14] + bat_19.iloc[q, 12] + bat_19.iloc[q, 8]) /
           (bat_19.iloc[q, 1] - bat_19.iloc[q, 4] + bat_19.iloc[q, 9] + bat_19.iloc[q, 16])
           for q in range(0, len(bat_19))]
bat_19['TA'] = TA_list
bat_19
```

### RC
빌 제임스가 고안한 한 타자가 (어떤 방식으로든) 만들어낸 득점 생산력. 당연히 높을수록 팀의 승리에 기여하는 바가 크다. 각 숫자들은 득점에 각 항목이 어느 정도 기여했는지를 알려주는 것이라 이해하면 된다. 예를 들면 도루가 0.62라는 것은 도루 한 개가 0.62점에 해당한다는 뜻.

```python
A: 안타 + 볼넷 - 도루실패 + 몸에맞는공 - 병살타
B: (1.125×단타) + (1.69 x 2루타) + (3.02×3루타) + (3.73×홈런) + .29×(BB-IBB+HBP) + .492×(SH+SF+SB) - (0.4×K)
C: 타수 + 볼넷 + 몸에맞는 공 + 희생번트 + 희생플라이
((2.4C + A)*(3C + B)) / 9C - 0.9C
추가사항 + 득점권 안타 수 - (득점권 타수 * 타율) + 주자 있을 때 홈런 - (주자 있을 때 타수 * 홈런) / 타수
    주자 있을 때 홈런이랑 타수를 어떻게 알아내냐고??????
```

```python
bat_19[bat_19.index == '나종덕']
```

```python
RC_list = []
for r in range(0, len(bat_19)) :
    A = HIT[r] + BB[r] - CS[r] + HP[r] - GD[r]
    hit1 = (bat_19.iloc[r, 4] - (bat_19.iloc[r, 5] + bat_19.iloc[r, 6] + bat_19.iloc[r, 7]))
    B = ((1.125 * hit1) + (1.69 * HIT2[r]) + (3.02 * HIT3[r]) + (3.73 * HR[r]) + 
         (0.29 * (BB[r] - IB[r] + HP[r])) + (0.492 * (SH[r] + SF[r] + SB[r])) - (0.4 * KK[r]))
    C = AB[r] + BB[r] + HP[r] + SH[r] + SF[r]
    RC_list.append((((2.4 * C) + A) * ((3 * C) + B)) / (9 * C) - (0.9 * C))
```

```python
bat_19['RC'] = RC_list
bat_19
```

### XR
RC보다 복잡하지만 좀 더 정확한 스탯으로 알려져 있다.

```python
1루타×0.5 + 2루타×0.72 + 3루타×1.04 + 홈런×1.44 + (몸에 맞는 볼 + 볼넷 - 고의4구)×0.34 +
고의4구×0.25 + 도루×0.18 - 도루실패×0.32 - (타수 - 안타 - 삼진)×0.09 - 삼진×0.098 -
병살타×0.37 + 희생플라이×0.37 + 희생번트×0.04.
```

```python
XR_list = []
for s in range(0, len(bat_19)) :
    hit1 = (bat_19.iloc[s, 4] - (bat_19.iloc[s, 5] + bat_19.iloc[s, 6] + bat_19.iloc[s, 7]))
    XR_list.append(((hit1 * 0.5) + (HIT2[s] * 0.72) + (HIT3[s] * 1.04) + (HR[s] * 1.44) + 
          ((HP[s] + BB[s] - IB[s]) * 0.34) + (IB[s] * 0.25) + (SB[s] * 0.18) - (CS[s] * 0.32) -
          ((AB[s] - HIT[s] - KK[s]) * 0.09) - (KK[s] * 0.098) - (GD[s] * 0.37) +
          (SF[s] * 0.37) + (SH[s] * 0.04)))
```

```python
bat_19['XR'] = XR_list
bat_19
```

### OW%
Offensive winning percentage. The theoretical winning percentage of a team comprising nine of the same players

```python
OW% = (RC/27)^2 / ((RC/27)^2 + 리그(RC/27)^2)
```

```python
os.chdir(r'C:\Users\jaij2\OneDrive\바탕 화면\논문\야구데이터\전처리')
bat_19 = pd.read_csv('2019_타자_Saber.csv', encoding = 'cp949')
bat_19
```

```python
RC
A: 안타 + 볼넷 - 도루실패 + 몸에맞는공 - 병살타
B: (1.125×단타) + (1.69 x 2루타) + (3.02×3루타) + (3.73×홈런) + .29×(BB-IBB+HBP) + .492×(SH+SF+SB) - (0.4×K)
C: 타수 + 볼넷 + 몸에맞는 공 + 희생번트 + 희생플라이
((2.4C + A)*(3C + B)) / 9C - 0.9C
```

```python
## 리그 RC/27
A_A = sum(HIT)/len(HIT) + sum(BB)/len(HIT) - sum(CS)/len(HIT) + sum(HP)/len(HIT) - sum(GD)/len(HIT)
A_B = ((1.125 * (sum(HIT)/len(HIT) - sum(HIT2)/len(HIT) - sum(HIT3)/len(HIT) - sum(HR)/len(HIT))) + 
       (1.69 * (sum(HIT2)/len(HIT))) + (3.02 * (sum(HIT3)/len(HIT))) + (3.73 * (sum(HR)/len(HIT))) + 
       (0.29 * (sum(BB)/len(HIT) - sum(IB)/len(HIT) + sum(HP)/len(HIT))) +
       (0.492 * (sum(SH)/len(HIT) + sum(SF)/len(HIT) + sum(SB)/len(HIT))) - (0.4 * (sum(KK)/len(HIT))))
A_C = sum(PA)/len(HIT) + sum(BB)/len(HIT) + sum(HP)/len(HIT) + sum(SH)/len(HIT) + sum(SF)/len(HIT)
A_RC = ((2.4 * A_C + A_A) * (3 * A_C + A_B)) / (9 * A_C) - 0.9 * A_C
A_RC = A_RC/27
A_RC
```

```python
OW_list = []
for z in bat_19['RC27'] :
    OW_list.append((z**2) / ((z**2) + (A_RC**2)))
```

```python
bat_19['OW%'] = OW_list
bat_19
```
