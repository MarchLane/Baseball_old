```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
os.chdir(r'C:\Users\jaij2\OneDrive\바탕 화면\논문\야구데이터\전처리')
bat_18 = pd.read_csv('2018_타자_DEA_OPS.csv', encoding = 'cp949')
bat_18 = bat_18.sort_values(by = ['Name'])
bat_18.index = bat_18['Name']
del bat_18['Name']
bat_18
```

```python
## 타석수, 출루율, 장타율, 18연봉을 칼럼으로
## 타석 수 144 이상 필터
bat_18 = bat_18[['PA', 'OBP', 'SLG', 'Money_18']]
bat_18.columns = ['PA_18', 'OBP_18', 'SLG_18', 'Money_18']
bat_18 = bat_18[bat_18['PA_18'] >= 144]
```

```python
bat_19 = pd.read_csv('2019_타자_DEA_OPS.csv', encoding = 'cp949')
bat_19 = bat_19.sort_values(by = ['Name'])
bat_19.index = bat_19['Name']
del bat_19['Name']
bat_19 = bat_19.drop(['Money_20'], axis = 1)
bat_19
```

```python
## 타석수, 출루율, 장타율, 19연봉을 칼럼으로
## 타석 수 144 이상 필터
bat_19 = bat_19[['PA', 'OBP', 'SLG', 'Money_19']]
bat_19.columns = ['PA_19', 'OBP_19', 'SLG_19', 'Money_19']
bat_19 = bat_19[bat_19['PA_19'] >= 144]
```

```python
## 18, 19년도 둘 다 있는 선수만 리스트로
ply_list = []
for a in bat_19.index :
    if a in bat_18.index :
        ply_list.append(a)
len(ply_list)
```

```python
## 18, 19년도 둘 중 하나라도 없는 선수들 제거하기
drop_list_1 = []
for c in range(0, len(bat_18)) :
    if bat_18.index[c] not in ply_list :
        drop_list_1.append(bat_18.index[c])
        
drop_list_2 = []
for d in range(0, len(bat_19)) :
    if bat_19.index[d] not in ply_list :
        drop_list_2.append(bat_19.index[d])
        
bat_18 = bat_18.drop(index = drop_list_1)
bat_19 = bat_19.drop(index = drop_list_2)
```

```python
bat_1918 = pd.DataFrame()
bat_1918.index = bat_18.index
bat_1918['PA_18'] = bat_18['PA_18']
bat_1918['PA_19'] = bat_19['PA_19']
bat_1918['OBP_18'] = bat_18['OBP_18']
bat_1918['OBP_19'] = bat_19['OBP_19']
bat_1918['SLG_18'] = bat_18['SLG_18']
bat_1918['SLG_19'] = bat_19['SLG_19']
bat_1918['Money_18'] = bat_18['Money_18']
bat_1918['Money_19'] = bat_19['Money_19']
bat_1918
```

```python
bat_19_18 = pd.DataFrame(index = bat_1918.index)
bat_19_18['PA19-18'] = [bat_1918.iloc[e, 1] - bat_1918.iloc[e, 0] for e in range(0, len(bat_1918))]
bat_19_18['OBP19-18'] = [bat_1918.iloc[e, 3] - bat_1918.iloc[e, 2] for e in range(0, len(bat_1918))]
bat_19_18['SLG19-18'] = [bat_1918.iloc[e, 5] - bat_1918.iloc[e, 4] for e in range(0, len(bat_1918))]
bat_19_18['Money19-18'] = [bat_1918.iloc[e, 7] - bat_1918.iloc[e, 6] for e in range(0, len(bat_1918))]
bat_19_18
```

```python
bat_19_18.to_csv('FA제외할라고temp.csv', encoding = 'cp949')
```
