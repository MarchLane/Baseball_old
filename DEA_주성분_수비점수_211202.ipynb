```python
pca = PCA(n_components = 3)
PC = pca.fit_transform(saber_3)
principals = pd.DataFrame(data = PC, columns = ['PC1', 'PC2', 'PC3'])
print(pd.DataFrame(pca.components_))
print('주성분 설명력', pca.explained_variance_ratio_)
print('주성분 설명력 합', sum(pca.explained_variance_ratio_))
print('고유값', pca.explained_variance_)
prins = pd.DataFrame(pca.components_)
prins.columns = saber_3.columns
prins.index = ['P1', 'P2', 'P3']
prins
```

```python
GEN_list = []
POW_list = []
SPD_list = []

for a in range(0, len(saber_3)) :
    GEN = 0
    POW = 0
    SPD = 0
    
    for b in range(0, len(saber_3.columns)) :
        GEN += saber_3.iloc[a, b] * prins.iloc[0, b]
        POW += saber_3.iloc[a, b] * prins.iloc[1, b]
        SPD += saber_3.iloc[a, b] * prins.iloc[2, b]
    
    GEN_list.append(GEN)
    POW_list.append(POW)
    SPD_list.append(SPD)
```
