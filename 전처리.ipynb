### OPS 계산

#### 출루율 (안타+볼넷+몸에 맞은 공) / (타수+볼넷+몸에 맞은 공+희생플라이)
(HIT+BB+HP)/(AB+BB+HP+SF)

```python
bat_19_DEA
```

```python
OBP_list = []
for i in range(0, len(bat_19_DEA)) :
    up = bat_19_DEA.iloc[i, 4] + bat_19_DEA.iloc[i, 12] + bat_19_DEA.iloc[i, 14]
    down = bat_19_DEA.iloc[i, 1] + bat_19_DEA.iloc[i, 12] + bat_19_DEA.iloc[i, 14] + bat_19_DEA.iloc[i, 11]
    OBP = up/down
    OBP_list.append(OBP)
```

```python
bat_19_DEA['OBP'] = OBP_list
bat_19_DEA
```

#### 장타율 (단타 수 + 2루타 수*2 + 3루타 수*3 + 홈런 수*4)/타수

```python
SLG_list = []
for j in range(0, len(bat_19_DEA)) :
    hit1 = bat_19_DEA.iloc[j, 4] - bat_19_DEA.iloc[j, 5] - bat_19_DEA.iloc[j, 6] - bat_19_DEA.iloc[j, 7]
    hit2 = bat_19_DEA.iloc[j, 5] * 2
    hit3 = bat_19_DEA.iloc[j, 6] * 3
    hit4 = bat_19_DEA.iloc[j, 7] * 4
    
    up = hit1 + hit2 + hit3 + hit4
    down = bat_19_DEA.iloc[j, 1]
    
    SLG = up/down
    SLG_list.append(SLG)
```

```python
bat_19_DEA['SLG'] = SLG_list
bat_19_DEA
```

```python
bat_19_DEA['OPS'] = [k + l for k, l in zip(bat_19_DEA['OBP'], bat_19_DEA['SLG'])]
bat_19_DEA
```
