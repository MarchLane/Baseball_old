```python
## bat_19(타격지표) 선수이름이랑 df_1(수비지표) 선수이름 매칭시켜서 팀명 넣기
t_list = []
for e in range(0, len(bat_19)) :
    for f in range(0, len(df_1)) :
        if bat_19.iloc[e, 0] == df_1.index[f] :
            t_list.append(df_1.iloc[f, 5])
```
