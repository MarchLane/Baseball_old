```python
OBP_list = []
for a in range(0, len(df_2)) :
    ja = (df_2['HIT'][a] + df_2['BB'][a] + df_2['HP'][a])
    mo = (df_2['AB'][a] + df_2['BB'][a] + df_2['HP'][a] + df_2['SF'][a])
    obp = ja/mo
    OBP_list.append(obp)
```

```python
SLG_list = []
for j in range(0, len(df_2)) :
    hit1 = df_2.iloc[j, 4] - df_2.iloc[j, 5] - df_2.iloc[j, 6] - df_2.iloc[j, 7]
    hit2 = df_2.iloc[j, 5] * 2
    hit3 = df_2.iloc[j, 6] * 3
    hit4 = df_2.iloc[j, 7] * 4
    
    up = hit1 + hit2 + hit3 + hit4
    down = df_2.iloc[j, 1]
    
    SLG = up/down
    SLG_list.append(SLG)
```

```python
df_fin['OBP'] = OBP_list
df_fin['SLG'] = SLG_list
df_fin['OPS'] = [o + s for o, s in zip(OBP_list, SLG_list)]
df_fin['SB'] = df_2['SB']
df_fin
```
