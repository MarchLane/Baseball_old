```python
os.chdir(r'C:\Users\jaij2\OneDrive\바탕 화면\논문\야구데이터\전처리')
df_SM = pd.read_csv('2019_타자_팀별_Saber.csv', encoding = 'cp949')
Name_list = df_SM['Unnamed: 0']
df_SM.index = Name_list
#df_SM = df_SM[df_SM['PA'] >= math.trunc(144 * 3.1)]
Money_20 = df_SM['Money_20']
df_SM = df_SM.iloc[:, -14:]
#df_SM['SECA'] = [int(e.replace('.', '').strip()) / 100000 for e in df_SM['SECA']]
# df_SM['RC'] = [int(e.replace('.', '').strip()) / 100000 for e in df_SM['RC']]
# df_SM['RC27'] = [int(e.replace('.', '').strip()) / 100000 for e in df_SM['RC27']]
# df_SM['XR'] = [int(e.replace('.', '').strip()) / 100000 for e in df_SM['XR']]
df_SM
```

```python
GEN_list = []
POW_list = []
SPD_list = []

for a in range(0, len(df_SM)) :
    GEN = 0
    POW = 0
    SPD = 0
    
    for b in range(0, len(df_SM.columns)) :
        GEN += df_SM.iloc[a, b] * df_SM_P.iloc[0, b]
        POW += df_SM.iloc[a, b] + df_SM_P.iloc[1, b]
        SPD += df_SM.iloc[a, b] * df_SM_P.iloc[2, b]
    
    GEN_list.append(GEN)
    POW_list.append(POW)
    SPD_list.append(SPD)
```

```python
df_prins = pd.DataFrame(index = df_SM.index)
df_prins['P1'] = GEN_list
df_prins['P2'] = POW_list
df_prins['P3'] = SPD_list
df_prins
```
