```python
## 선수 한 명의 1년치 기록 합치기
test = 0
ID_1 = 60100
row = 0
for d in range(1, len(bat_19.columns)) :
    c = 0
    row = 0
    for a in range(0, len(bat_19)) :
        ID = bat_19.iloc[a, 0]
        if ID == ID_1 :
            c += bat_19.iloc[a, d]

        else :
            bat_19_sum.iloc[row, d] = c
            row += 1
            c = 0
            c += bat_19.iloc[a, d]
            ID_1 = bat_19.iloc[a, 0]
            test += 1
```

```python
## 연봉 추가
os.chdir(r'C:\Users\jaij2\OneDrive\바탕 화면\논문\야구데이터\2020빅콘테스트 문제데이터(데이터분석분야-퓨처스리그)\2020빅콘테스트 데이터분석분야-퓨쳐스리그_스포츠투아이_제공데이터(.CSV)_시즌별, 시트별 구분')
ply_19 = pd.read_csv('2020빅콘테스트_스포츠투아이_제공데이터_선수_2019.csv', encoding = 'cp949')
ply_19.index = ply_19['PCODE']
ply_19
```

```python
## 선수명, 포지션, 연봉 추가
bat_19_sum['Name'] = ply_19['NAME']
bat_19_sum['Position'] = ply_19['POSITION']
bat_19_sum['Salary'] = ply_19['MONEY']
bat_19_sum['Salary'] = [g[:-2] for g in bat_19_sum['Salary'].to_list()]
bat_19_sum['Team'] = ply_19['T_ID']
bat_19_sum['Age'] = ply_19['AGE_VA']
bat_19_sum
```

```python
## 투수 빼기
bat_19_sum = bat_19_sum[bat_19_sum['Position'] != '투']
bat_19_sum
```

```python
## DEA 돌릴 df 생성
bat_19_DEA = bat_19_sum.reset_index(drop = True)
bat_19_DEA.index = bat_19_sum['Name']
del bat_19_DEA['Name']
bat_19_DEA
```

```python
OBP_list = []
for i in range(0, len(bat_19_DEA)) :
    up = bat_19_DEA.iloc[i, 4] + bat_19_DEA.iloc[i, 12] + bat_19_DEA.iloc[i, 14]
    down = bat_19_DEA.iloc[i, 1] + bat_19_DEA.iloc[i, 12] + bat_19_DEA.iloc[i, 14] + bat_19_DEA.iloc[i, 11]
    OBP = up/down
    OBP_list.append(OBP)
```

```python
bat_19_DEA['OBP'] = OBP_list
bat_19_DEA
```

```python
SLG_list = []
for j in range(0, len(bat_19_DEA)) :
    hit1 = bat_19_DEA.iloc[j, 4] - bat_19_DEA.iloc[j, 5] - bat_19_DEA.iloc[j, 6] - bat_19_DEA.iloc[j, 7]
    hit2 = bat_19_DEA.iloc[j, 5] * 2
    hit3 = bat_19_DEA.iloc[j, 6] * 3
    hit4 = bat_19_DEA.iloc[j, 7] * 4
    
    up = hit1 + hit2 + hit3 + hit4
    down = bat_19_DEA.iloc[j, 1]
    
    SLG = up/down
    SLG_list.append(SLG)
```

```python
bat_19_DEA['SLG'] = SLG_list
bat_19_DEA
```

```python
bat_19_DEA['OPS'] = [k + l for k, l in zip(bat_19_DEA['OBP'], bat_19_DEA['SLG'])]
bat_19_DEA
```
