```python
## 144타석 이상
df_21 = df_2[df_2['PA'] >= 144]
## 신인 이학주, 노시환 제외
df_21 = df_21.drop(['이학주', '노시환'])
df_21
```

### 주성분

```python
df_22 = df_21.iloc[:, -14:]
df_22
```

```python
pca = PCA(n_components = 3)
PC = pca.fit_transform(df_22)
principals = pd.DataFrame(PC, columns = ['P1', 'P2', 'P3'])
print('주성분 설명력 : ', pca.explained_variance_ratio_)
print('주성분 설명력 합 :', sum(pca.explained_variance_ratio_))
print('고유값', pca.explained_variance_)
prins = pd.DataFrame(pca.components_)
prins.columns = df_22.columns
prins.index = ['P1', 'P2', 'P3']
prins
```

```python
GEN_list = []
POW_list = []
SPD_list = []

for a in range(0, len(df_22)) :
    GEN = 0
    POW = 0
    SPD = 0
    
    for b in range(0, len(df_22.columns)) :            
        GEN += df_22.iloc[a, b] * prins.iloc[0, b]
        POW += df_22.iloc[a, b] * prins.iloc[1, b]
        SPD += df_22.iloc[a, b] * prins.iloc[2, b]
    
    GEN_list.append(GEN)
    POW_list.append(POW)
    SPD_list.append(SPD)
```

```python
df_fin = pd.DataFrame(index = df_22.index)
df_fin['P1'] = GEN_list
df_fin['P2'] = POW_list
df_fin['P3'] = SPD_list
df_fin
```

```python
df_fin['E/IP'] = 1 - df_1['E'] / df_1['IP']
df_fin['Sal_19'] = df_21['Money_19']
df_fin
```
