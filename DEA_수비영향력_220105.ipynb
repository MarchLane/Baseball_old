```python
import os
import pandas as pd
from statsmodels.formula.api import ols
from sklearn.decomposition import PCA
import numpy as np
```

```python
os.chdir(r'C:\Users\jaij2\OneDrive\바탕 화면\논문\야구데이터\전처리')
df_1 = pd.read_csv('2019_타자_Saber.csv', encoding = 'cp949')
df_1.index = df_1['Name']
del df_1['Name']
df_1
```

```python
## 144타석 이상
df_11 = df_1[df_1['PA'] >= 144]
df_11
## 김준태 제외 (PSN이 nan값)
# df_11 = df_11.drop('김준태')
# df_11
```

```python
## 신인 제외
df_11 = df_11.drop(['이학주', '노시환'])
df_11
```

```python
## Sabermetrics 지표만
df_12 = df_11.iloc[:, -14:]
df_12
```

```python
pca = PCA(n_components = 3)
PC = pca.fit_transform(df_12)
principals = pd.DataFrame(PC, columns = ['PC1', 'PC2', 'PC3'])
print('주성분 설명력 : ', pca.explained_variance_ratio_)
print('주성분 설명력 합 :', sum(pca.explained_variance_ratio_))
print('고유값', pca.explained_variance_)
prins = pd.DataFrame(pca.components_)
prins.columns = df_12.columns
prins.index = ['P1', 'P2', 'P3']
prins
```

```python
GEN_list = []
POW_list = []
SPD_list = []

for a in range(0, len(df_12)) :
    GEN = 0
    POW = 0
    SPD = 0
    
    for b in range(0, len(df_12.columns)) :              ## 주성분 설명력 추가 -> 근데 결과 차이없음
        GEN += df_12.iloc[a, b] * prins.iloc[0, b] * pca.explained_variance_ratio_[0]
        POW += df_12.iloc[a, b] * prins.iloc[1, b] * pca.explained_variance_ratio_[1]
        SPD += df_12.iloc[a, b] * prins.iloc[2, b] * pca.explained_variance_ratio_[2]
    
    GEN_list.append(GEN)
    POW_list.append(POW)
    SPD_list.append(SPD)
```

```python
df_fin = pd.DataFrame(index = df_12.index)
df_fin['P1'] = GEN_list
df_fin['P2'] = POW_list
df_fin['P3'] = SPD_list
df_fin
```

```python
## 연봉, 로그연봉 추가
df_fin['Sal'] = df_11['Money_19']
df_fin['LogSal'] = np.log(df_11['Money_19'])
df_fin
```

```python
e_list = []
for d in range(0, len(df_4)) :
    for e in range(0, len(df_3)) :
        if df_3.index[e] == df_4.index[d] :
            e_list.append(df_4.iloc[d, 5])
```

```python
df_fin['Exp'] = df_3['연차']
df_fin['FA'] = df_3['FA유무']
df_fin['Erate'] = e_list
df_fin['E'] = df_4['E']
df_fin['IP'] = df_4['IP']
df_fin['Pos'] = df_3['Position']
df_fin
```

```python
out_list = []
in_list = []
cat_list = []
for f in range(0, len(df_fin['Pos'])) :
    if df_fin['Pos'][f] == '외' :
        out_list.append(1 * df_fin['IP'][f])
    else : 
        out_list.append(0)
        
    if df_fin['Pos'][f] == '내' :
        in_list.append(1.5 * df_fin['IP'][f])
    else :
        in_list.append(0)
        
    if df_fin['Pos'][f] == '포' :
        cat_list.append(2 * df_fin['IP'][f])
    else : 
        cat_list.append(0)
        
```

```python
df_fin['catcher'] = cat_list
df_fin['infielder'] = in_list
df_fin['outfielder'] = out_list
df_fin['pos_score'] = df_fin['catcher'] + df_fin['infielder'] + df_fin['outfielder']
df_fin
```

```python
model = ols(formula = 'LogSal ~ P1 + Exp + FA + pos_score', data = df_fin).fit()
```
