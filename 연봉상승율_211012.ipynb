```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
os.chdir(r'C:\Users\jaij2\OneDrive\바탕 화면\논문\야구데이터\전처리')
bat_19 = pd.read_csv('2019_타자_Saber.csv', encoding = 'cp949')
bat_19.index = bat_19['Name']
del bat_19['Name']
bat_19 = bat_19[bat_19['PA'] >= 144]
bat_19 = bat_19.drop(['Position'], axis = 1)
bat_19
```

```python
y : 타석 수, 출루율, 장타율
x : 19 -> 20 상승률 or 상승폭 
DEA 에서 연봉 동결이나 하락한 경우 투입이 작기 때문에 가장 효율적인 것으로 나옴
-> 연봉하락이나 동결한 경우 제외하고 다시 분석
```

```python
bat_19['연봉상승률'] = [round((bat_19['Money_20'][a] / bat_19['Money_19'][a]) -1, 2)  
                   for a in range(0, len(bat_19))]
bat_19['연봉상승폭'] = [bat_19['Money_20'][b] - bat_19['Money_19'][b] 
                   for b in range(0, len(bat_19))]
bat_19
```

```python
bat_19 = bat_19[bat_19['연봉상승률'] > 0]
bat_19
```

```python
bat_19.to_csv('2019_타자_연봉상승.csv', encoding = 'cp949')
```

```python
scaler = MinMaxScaler()
bat_19_mm = scaler.fit_transform(bat_19)
bat_19_mm = pd.DataFrame(bat_19_mm)
bat_19_mm.columns = bat_19.columns
bat_19_mm.index = bat_19.index
bat_19_mm
```
